```python
import feedparser
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import nltk
from bs4 import BeautifulSoup

# Download the VADER lexicon before the first use
nltk.download('vader_lexicon')

# RSS link
rss_url = "https://cointelegraph.com/rss"

# Parsing the RSS Feed
feed = feedparser.parse(rss_url)

# Initialize the VADER sentiment analyzer
sid = SentimentIntensityAnalyzer()

# Sentiment analysis for each article
for entry in feed.entries:
    if 'summary' in entry:
        html_content = entry.summary
    else:
        html_content = entry.title  # Fallback to title if no full content is available

    # Use BeautifulSoup to extract text from HTML
    text_content = BeautifulSoup(html_content, 'html.parser').get_text()

    sentiment_score = sid.polarity_scores(text_content)
    print(f"Article Text: {text_content[:100]}...")  # Print the first 100 characters of the article text
    print(f"Sentiment Score: {sentiment_score}\n")
```

```python
import feedparser
import requests
from bs4 import BeautifulSoup
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import nltk

# Download the VADER lexicon before the first use
nltk.download('vader_lexicon')

# RSS link
rss_url = "https://cointelegraph.com/rss"

# Parsing the RSS Feed
feed = feedparser.parse(rss_url)

# Initialize the VADER sentiment analyzer
sid = SentimentIntensityAnalyzer()

# Sentiment analysis for each article
for entry in feed.entries:
    article_url = entry.link  # 获取文章链接

    # 获取文章页面内容
    response = requests.get(article_url)
    html_content = response.content

    # 使用BeautifulSoup提取文本内容
    soup = BeautifulSoup(html_content, 'html.parser')
    article_text = soup.get_text()

    # 对文章文本进行情感分析
    sentiment_score = sid.polarity_scores(article_text)
    print(f"Article URL: {article_url}")
    print(f"Sentiment Score: {sentiment_score}\n")
```
